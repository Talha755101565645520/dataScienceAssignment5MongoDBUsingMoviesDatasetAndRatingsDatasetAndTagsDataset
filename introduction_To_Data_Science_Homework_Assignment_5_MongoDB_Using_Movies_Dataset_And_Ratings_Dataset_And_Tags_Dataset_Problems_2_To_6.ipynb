```python
############################################################################################################################
# Problem 3: Write a Program to read the given 3 different Commas-Separated Value Files that includes the Commas-Separated #
# Value CSV File "movies.csv", the Commas-Separated Value CSV File "ratings.csv", and the Commas-Separated Value CSV       #
# File "tags.csv" and then insert all the records into three different collections.                                        #                                                                   
############################################################################################################################
# Create a connection to MongoDB.
client = MongoClient('mongodb://localhost:27017')

# Create a database in MongoDB.
db = client['movie_database']

# Create the three different collections.
movies_collection = db['movies']
ratings_collection = db['ratings']
tags_collection = db['tags']

# Use the "delete_many()" Function for the three different collections to clean out any data already in the three different collections so that 
# they are empty every run.
movies_collection.delete_many({})
ratings_collection.delete_many({})
tags_collection.delete_many({})

# Read the Commas-Separated Value CSV Files "movies.csv", "ratings.csv", and "tags.csv" using the Python Library "pandas".
movies_data_frame = pd.read_csv('movies.csv')
ratings_data_frame = pd.read_csv('ratings.csv')
tags_data_frame = pd.read_csv('tags.csv')

# Convert the Dataframes to a Dictionary using the "to_dict()" Function and use the "insert_many()" Functin to insert 
# into the collections.
movies_collection.insert_many(movies_data_frame.to_dict('records'))
ratings_collection.insert_many(ratings_data_frame.to_dict('records'))
tags_collection.insert_many(tags_data_frame.to_dict('records'))
print("Insert successful!")
```

```python
##################################################################################
# Problem 6 Part a: Develop code to find the number of movies released per year. #
##################################################################################

# Use an Aggregation Pipeline consisting of a project, group, and sort aggregate in order to produce a list of "years" and the number of movies released
# in those "years".
years = movies_collection.aggregate([
    {"$project": {
        "year": {
            "$regexFind": {"input": "$title", "regex": "\\(([0-9]{4})\\)"}
        }
    }},
    {"$group": {"_id": "$year.captures", "count": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])
list(years)
```

```python
##########################################################################
# Problem 6 Part b: Develop code to find the number of movies per genre. #
##########################################################################

# Use an Aggregation Pipeline consisting of a project and a split, unwind, group and sort aggregate in order to produce a list of "genres" and 
# compute the number of movies for each different type of genre. First, use the "project" aggregate and "split" aggregate to split the genres 
# on the '|' character, in other words, the Pipe Character. Second, use the "unwind" aggregate to split the array into separate documents. Third,
# use the "group" aggregate to create the count. Finally, use the "sort" aggregate to show the results for each "genre" type category in 
# alphabetical order. 
genre = movies_collection.aggregate([
    {"$project": {
        "genre": {
            "$split": ["$genres", "|"]
        }
    }},
    {"$unwind": "$genre"},
    {"$group": {"_id": "$genre", "count": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])
list(genre)
```

```python
###########################################################################
# Problem 6 Part c: Develop code to find the number of movies per rating. #
###########################################################################

# Use an Aggregation Pipeline consisting of a group aggregate to count the number of entries for each rating value and the sort aggregate to sort
# each rating value from the lowest rating value to the highest rating value.
rating = ratings_collection.aggregate([
    {"$group": {"_id": "$rating", "count": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])
list(rating)
```

```python
#######################################################################
# Problem 6 Part d: Develop code to find the number of movies tagged. #
#######################################################################

# Use an Aggregation Pipeline by first using one "group" aggregate to group the movies by "movieId" to produce a set of unique movies and then 
# using a "group" aggregate to group by "None" so that the group would count all the lines as the same key value.
tag = tags_collection.aggregate([
    {"$group": {"_id": "$movieId", "count": {"$sum": 1}}},
    {"$group": {"_id": None, "count": {"$sum": 1}}}
])
list(tag)
```

```python
################################################################
# Problem 6 Part e: Develop code to find the most popular tag. #
################################################################

# Use an Aggregation Pipeline consisting of a group, sort, and limit aggregate where first a "group" aggregate is used to count all of the tags,
# second a "sort" aggregate is used to sort the count for each different "tag" from the most popular tag to the least popular tag, and then
# finally, a "limit" aggregate is used to set the limit to the single most popular tag with the highest count value.
tag = tags_collection.aggregate([
    {"$group": {"_id": "$tag", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 1}
])
list(tag)
```
